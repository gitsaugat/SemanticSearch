```python
from huggingface_hub import notebook_login
from sentence_transformers import SentenceTransformer, util

# model to use
# all-MiniLM-L6-v2
```

```python
query = "How to schedule a physician visit?"

corpus = [
    "How can I book a doctor appointment?",
    "What insurance plans are accepted?",
    "Where can I find my lab test results?",
    "How do I update my personal information?"
]
```

```python
model = SentenceTransformer('all-MiniLM-L6-v2')
```

```python
query_embedding = model.encode(query, convert_to_tensor=True)
corpus_embeddings = model.encode(corpus, convert_to_tensor=True)
```

```python
query_embedding.shape
```

```python
corpus_embeddings.shape
```

```python
cosine_similarity_score = util.cos_sim(query_embedding, corpus_embeddings)
```

```python
cosine_similarity_score
```

```python
best_similarity_index = cosine_similarity_score.argmax()
```

```python
best_similarity_index
```

```python
best_score= cosine_similarity_score[0][best_similarity_index]
best_score
```

```python
best_match_sentence = corpus[best_similarity_index]
```
